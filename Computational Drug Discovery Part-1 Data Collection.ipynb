### Importing Libraries


```python
import pandas as pd
from chembl_webresource_client.new_client import new_client
```

### Search for Target Protein

#### Target search for coronavirus

```python
# Target search for coronavirus
target = new_client.target
target_query = target.search('coronavirus')
targets = pd.DataFrame.from_dict(target_query)
targets
```

```python
type(target_query)
```

### Select and retrieve bioactivity data for SARS coronavirus 3C-like proteinase(5th entry)

```python
selected_target = targets.target_chembl_id[4]
selected_target
```

Only bioactivity data for *SARS coronavirus 3C-like proteinase* (CHEMBL3927) that are reported as IC50 values are retrieved as it will make it more uniform and we won't have a mixture of different bioactivity units

```python
activity = new_client.activity
res = activity.filter(target_chembl_id = selected_target).filter(standard_type = "IC50")
```

```python
df = pd.DataFrame.from_dict(res)
```

### Handling Missing Values

If any compound has missing value for the standard_value column, it is dropped.

```python
df2 = df[df.standard_value.notna()]
df2.head()
```

For this dataset, there is no missing data.

### Data pre-processing of the bioactivity data


#### Combine the 3 columns (molecule_chembl_id,canonical_smiles,standard_value) and bioactivity_class into a DataFrame

```python
selection = ['molecule_chembl_id', 'canonical_smiles', 'standard_value']
df3 = df2[selection]
df3.head()
```

#### Labelling compunds as active, inactive or intermediate

The bioactivity data is in the IC50 unit.

Compounds are labelled into active, inactive or intermediate according to the following criteria:

* Active: Values less than 1000 nM 
* Inactive: Values exceeding 10,000 nM
* Intermediate: Values between 1000 nM and 10,000 nM

```python
bioactivity_class =[]
for i in df3.standard_value:
    if float(i) >= 10000:
        bioactivity_class.append("inactive")
    elif float(i) <= 1000:
        bioactivity_class.append("active")
    else:
        bioactivity_class.append("intermediate")      
```

```python
bioactivity_class = pd.Series(bioactivity_class, name='bioactivity_class')
df4 = pd.concat([df3, bioactivity_class], axis=1)
df4
```

```python
df4.to_csv("bioactivity_data_preprocessed.csv", index = False)
```
